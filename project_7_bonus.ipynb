```python
import pandas as pd
from scipy import stats
import seaborn as sns
sns.set_style('whitegrid')
```

```python
covid_df = pd.read_csv('day_wise.csv')
covid_df.describe()
#For this project I will be doing a linear regression fitting the number of deaths to the days past
```

```python
#Since the dataset is arranged by date I can use the the index my x, otherwise I would have to transform the data
lin_regress =  stats.linregress(x=covid_df.index, y=covid_df['Recovered / 100 Cases'])
lin_regress.slope, lin_regress.intercept
```

```python
#Gavin Albrecht
axes = sns.regplot(x=covid_df.index, y=covid_df['Recovered / 100 Cases'])
```

```python
recovery_2020_07_28 = 188*lin_regress.slope+ lin_regress.intercept
print(recovery_2020_07_28)
```
