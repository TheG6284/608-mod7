```python
nyc = pd.read_csv('ave_hi_nyc_jan_1895-2018.csv')
```

```python
nyc.columns = ['Date', 'Temperature', 'Anomaly']
nyc.describe
```

```python
lin_regress =  stats.linregress(x=nyc.Date, y=nyc.Temperature)
print(lin_regress.slope, lin_regress.intercept)
```

```python
high_2021 = lin_regress.slope * 2021 +lin_regress.intercept
print(high_2021)
```
